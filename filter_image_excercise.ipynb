```python
def blur(img):
    avg_img = (img[1:-1,1:-1] + img[:-2,1:-1] + img[2:,1:-1] + img[1:-1,2:] +
               img[1:-1,:-2]) / 5.0
    return avg_img
```

```python
blurred = blur(img)
```

```python
plt.figure()
plt.gray()
plt.subplot(1,3,1)
plt.imshow(img)
plt.title('original')
plt.subplot(1,3,2)
plt.imshow(blurred)
plt.title('blurred image')
plt.subplot(1,3,3)
plt.imshow(img[1:-1,1:-1] - blurred)
plt.title('difference')
```

```python
for num in range(50):
    avg_img = blur(blurred)

# Plot the original image first
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('original')

# Now the filtered image.
plt.subplot(1,2,2)
plt.imshow(blurred)
plt.title('smoothed 50 times')

# assert np.allclose(smooth(img), smooth_loop(img))

plt.show()
```
